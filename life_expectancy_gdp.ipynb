```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm

data = pd.read_csv('all_data.csv')
print(data['Country'].unique())
```

```python
print(data.head())
```

```python
chile = data[data['Country']=='Chile']
china = data[data['Country']=='China']
germany = data[data['Country']=='Germany']
mexico = data[data['Country']=='Mexico']
usa = data[data['Country']=='United States of America']
zimbabwe = data[data['Country']=='Zimbabwe']

plt.subplot(2, 3, 1)
plt.scatter(chile['Life expectancy at birth (years)'], chile['GDP'])
plt.title('Chile')
plt.subplot(2, 3, 2)
plt.scatter(china['Life expectancy at birth (years)'], china['GDP'])
plt.title('China')
plt.subplot(2, 3, 3)
plt.scatter(germany['Life expectancy at birth (years)'], germany['GDP'])
plt.title('Germany')
plt.subplot(2, 3, 4)
plt.scatter(mexico['Life expectancy at birth (years)'], mexico['GDP'])
plt.title('Mexico')
plt.subplot(2, 3, 5)
plt.scatter(usa['Life expectancy at birth (years)'], usa['GDP'])
plt.title('USA')
plt.subplot(2, 3, 6)
plt.scatter(zimbabwe['Life expectancy at birth (years)'], zimbabwe['GDP'])
plt.title('Zimbabwe')

plt.tight_layout()

plt.show()
plt.clf()
```
